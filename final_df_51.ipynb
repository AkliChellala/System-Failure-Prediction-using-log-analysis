```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('testout51.csv')
```

```python
time=df['CPU 1 YBLPVDAKDLWAPP1']
```

```python
df.drop(['Idle%_CPU1','Idle%_CPU2','Idle%_CPU3','Idle%_CPU4','Idle%_CPU5'],axis=1,inplace=True)

df.drop(['Idle%_CPU6','Idle%_CPU7','Idle%_CPU8','Idle%_CPU9','Idle%_CPU10','Idle%_CPU11','Idle%_CPU12','Idle%_CPU13'],axis=1,inplace=True)

df.drop(['Idle%_CPU14','Idle%_CPU15','Idle%_CPU16','Idle%_CPU17'],axis=1,inplace=True)
```

```python
Cpu=np.log(np.sum(df.iloc[:,1:52],axis=1))
CPU=pd.DataFrame({'cpu':Cpu})
```

```python
DISK=pd.DataFrame({'disk':disk})

NETPACKET=pd.DataFrame({'netpacket':netpacket})

RAM=pd.DataFrame({'RAM':RAM})

df=pd.concat([CPU,DISK,NETPACKET,RAM],axis=1)
```

```python
class_label = pd.DataFrame(df.iloc[:,-1])
class_label.columns = ['label']
df = df.iloc[:, :-1]
df.head(n=10)
df = df.sub(df.mean(axis=0), axis=1)
df.head()
df_mat = np.asmatrix(df)
sigma = np.cov(df_mat.T)
eigVals, eigVec = np.linalg.eig(sigma)
sorted_index = eigVals.argsort()[::-1] 
eigVals = eigVals[sorted_index]
eigVec = eigVec[:,sorted_index]
eigVec = eigVec[:,:1]
transformed = df_mat.dot(eigVec)
#horizontally stack transformed data set with class label.
final_df = np.hstack((transformed, class_label))
#convert the numpy array to data frame
final_df = pd.DataFrame(final_df)
#define the column names
final_df.columns = ['x','label']
```

```python
final_df.drop(['label'],axis=1,inplace=True)
x=final_df['x']
features=pd.concat([time,x],axis=1)
features.head()
```

```python
features.to_csv('final_df_51.csv')
```
